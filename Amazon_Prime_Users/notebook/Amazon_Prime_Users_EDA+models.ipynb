### Model Building

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_auc_score
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC

from xgboost import XGBClassifier
from lightgbm import LGBMClassifier
from catboost import CatBoostClassifier

# Split data
X = df_model.drop(columns=['User_Status_Encoded'])
y = df_model['User_Status_Encoded']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, stratify=y, random_state=42)
```

```python
# Dictionary of models
models = {
    'Logistic Regression': LogisticRegression(max_iter=1000),
    'Decision Tree': DecisionTreeClassifier(),
    'KNN': KNeighborsClassifier(),
    'Naive Bayes': GaussianNB(),
    'SVM': SVC(probability=True),
    'Random Forest': RandomForestClassifier(),
    'XGBoost': XGBClassifier(use_label_encoder=False, eval_metric='logloss'),
    'LightGBM': LGBMClassifier(),
    'CatBoost': CatBoostClassifier(verbose=0)
}

results = []

for name, model in models.items():
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    y_proba = model.predict_proba(X_test)[:, 1]

    results.append({
        'Model': name,
        'Accuracy': accuracy_score(y_test, y_pred),
        'Precision': precision_score(y_test, y_pred),
        'Recall': recall_score(y_test, y_pred),
        'F1 Score': f1_score(y_test, y_pred),
        'ROC AUC': roc_auc_score(y_test, y_proba)
    })

# DataFrame of results
results_df = pd.DataFrame(results).sort_values(by='F1 Score', ascending=False)
print(results_df)
```

```python
plt.figure(figsize=(10,6))
sns.barplot(data=results_df.melt(id_vars='Model'), x='value', y='Model', hue='variable')
plt.title("Baseline Model Performance Comparison")
plt.xlabel("Score")
plt.ylabel("Model")
plt.legend(title="Metric", bbox_to_anchor=(1.05, 1))
plt.tight_layout()
plt.show()
```

```python
from sklearn.metrics import classification_report, confusion_matrix, ConfusionMatrixDisplay
```

```python
for name, model in models.items():
    print(f"\n{'='*30}\nModel: {name}\n{'='*30}")
    
    # Predict
    y_pred = model.predict(X_test)
    
    # 1️⃣ Classification Report
    print("Classification Report:")
    print(classification_report(y_test, y_pred, target_names=["Active (0)", "Churned (1)"]))
    
    # 2️⃣ Confusion Matrix
    cm = confusion_matrix(y_test, y_pred)
    disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=["Active", "Churned"])
    
    # Plot
    disp.plot(cmap='Blues')
    plt.title(f"Confusion Matrix: {name}")
    plt.tight_layout()
    plt.show()
```

```python
from sklearn.metrics import roc_curve, auc

plt.figure(figsize=(10, 8))

for name, model in models.items():
    # Predict probabilities
    y_proba = model.predict_proba(X_test)[:, 1]
    
    # Compute ROC
    fpr, tpr, _ = roc_curve(y_test, y_proba)
    roc_auc = auc(fpr, tpr)
    
    # Plot
    plt.plot(fpr, tpr, lw=2, label=f'{name} (AUC = {roc_auc:.2f})')

# Plot reference line (random classifier)
plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')

# Plot formatting
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curves for All Baseline Models')
plt.legend(loc='lower right')
plt.grid()
plt.tight_layout()
plt.show()
```

# Hyperparameters Tuning
#### LightGBM - RandomizedSearchCV

```python
from lightgbm import LGBMClassifier

lgbm = LGBMClassifier(random_state=42)

param_lgbm = {
    'num_leaves': [15, 31, 63],
    'max_depth': [-1, 5, 10, 15],
    'learning_rate': [0.01, 0.05, 0.1],
    'n_estimators': [100, 200, 500]
}

cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

rs_lgbm = RandomizedSearchCV(lgbm, param_distributions=param_lgbm,
                             n_iter=20, scoring='f1', cv=cv,
                             verbose=1, random_state=42, n_jobs=-1)

rs_lgbm.fit(X, y)
print("Best Params - LightGBM:", rs_lgbm.best_params_)
```

#### XGBoost: RandomizedSearchCV

```python
from xgboost import XGBClassifier

xgb = XGBClassifier(use_label_encoder=False, eval_metric='logloss', random_state=42)

param_xgb = {
    'max_depth': [3, 5, 10],
    'learning_rate': [0.01, 0.05, 0.1],
    'n_estimators': [100, 200, 500],
    'subsample': [0.6, 0.8, 1.0],
    'colsample_bytree': [0.6, 0.8, 1.0]
}

rs_xgb = RandomizedSearchCV(xgb, param_distributions=param_xgb,
                            n_iter=20, scoring='f1', cv=cv,
                            verbose=1, random_state=42, n_jobs=-1)

rs_xgb.fit(X, y)
print("Best Params - XGBoost:", rs_xgb.best_params_)
```

#### Decision Tree: GridSearchCV

```python
from sklearn.tree import DecisionTreeClassifier

dt = DecisionTreeClassifier(random_state=42)

param_dt = {
    'max_depth': [3, 5, 10, 15],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4]
}

gs_dt = GridSearchCV(dt, param_grid=param_dt,
                     scoring='f1', cv=cv,
                     verbose=1, n_jobs=-1)

gs_dt.fit(X, y)
print("Best Params - Decision Tree:", gs_dt.best_params_)
```

#### KNN: GridSearchCV

```python
from sklearn.neighbors import KNeighborsClassifier

knn = KNeighborsClassifier()

param_knn = {
    'n_neighbors': [3, 5, 7, 9, 11],
    'weights': ['uniform', 'distance'],
    'p': [1, 2]  # Manhattan (1) or Euclidean (2)
}

gs_knn = GridSearchCV(knn, param_grid=param_knn,
                      scoring='f1', cv=cv,
                      verbose=1, n_jobs=-1)

gs_knn.fit(X, y)
print("Best Params - KNN:", gs_knn.best_params_)
```

```python
models_tuned = {
    'LightGBM': rs_lgbm.best_estimator_,
    'XGBoost': rs_xgb.best_estimator_,
    'Decision Tree': gs_dt.best_estimator_,
    'KNN': gs_knn.best_estimator_
}
```

```python
from sklearn.metrics import (
    accuracy_score, precision_score, recall_score, f1_score,
    roc_auc_score, classification_report, confusion_matrix,
    ConfusionMatrixDisplay, roc_curve
)

# ✅ Dictionary of tuned models
models_tuned = {
    'LightGBM': rs_lgbm.best_estimator_,
    'XGBoost': rs_xgb.best_estimator_,
    'Decision Tree': gs_dt.best_estimator_,
    'KNN': gs_knn.best_estimator_
}

# ✅ Store all results
final_results = []

# ✅ ROC Plot Setup
plt.figure(figsize=(10, 8))

for name, model in models_tuned.items():
    # Train model
    model.fit(X_train, y_train)

    # Predictions
    y_pred = model.predict(X_test)
    y_proba = model.predict_proba(X_test)[:, 1]  # probability for class 1 (churn)

    # Evaluation metrics
    acc = accuracy_score(y_test, y_pred)
    prec = precision_score(y_test, y_pred)
    rec = recall_score(y_test, y_pred)
    f1 = f1_score(y_test, y_pred)
    auc_score = roc_auc_score(y_test, y_proba)

    # Append to results
    final_results.append({
        'Model': name,
        'Accuracy': acc,
        'Precision': prec,
        'Recall': rec,
        'F1 Score': f1,
        'ROC AUC': auc_score
    })

    # ================== 📊 CLASSIFICATION REPORT ==================
    print(f"\n{'='*40}\n📌 Model: {name}\n{'='*40}")
    print(classification_report(y_test, y_pred, target_names=["Active", "Churned"]))

    # ================== 📊 CONFUSION MATRIX ==================
    cm = confusion_matrix(y_test, y_pred)
    disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=["Active", "Churned"])
    disp.plot(cmap='Blues')
    plt.title(f"Confusion Matrix: {name}")
    plt.tight_layout()
    plt.show()

    # ================== 📈 ROC CURVE ==================
    fpr, tpr, _ = roc_curve(y_test, y_proba)
    plt.plot(fpr, tpr, label=f"{name} (AUC = {auc_score:.2f})")

# Plot ROC Curve All-in-One
plt.plot([0, 1], [0, 1], 'k--', label="Random Guess")
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.title("ROC Curve Comparison (Tuned Models)")
plt.legend(loc="lower right")
plt.grid()
plt.tight_layout()
plt.show()

# ================== 📋 FINAL COMPARISON TABLE ==================
final_results_df = pd.DataFrame(final_results).sort_values(by='F1 Score', ascending=False)
print("\n📊 Final Model Evaluation Summary:")
print(final_results_df)
```
